```python
def age_cat(x):
    if x<18:
        return "меньше 18"
    elif x<30:
        return "от 18 до 30"
    elif x>=30:
        return "30 и больше"
    else:
        return x
```

```python
df['Age_category'] = df['Age'].apply(age_cat)
df.head()
```

```python
df[df['Age_category'] == '30 и больше'].drop_duplicates('ID').shape[0]
```
